## Задача 1
Даны две независимые выборки.

Не соблюдается условие нормальности

x1 380,420, 290

y1 140,360,200,900

Сделайте вывод по результатам, полученным с помощью функции

```python
# Импортируем необходимые модули
import numpy as np
import scipy.stats as stats
```

```python
# Не соблюдается условие нормальности, необходимо применять непараметрические тесты.
# Выборки независимые и их всего 2 можно применить тест Манна-Уитни
x1 = np.array([380,420, 290])
y1 = np.array([140,360,200,900])
```

```python
# H0: μ1 = μ2 - Медианы первой и второй выборок принадлежат одной генеральной совокупности
# H0: μ1 ≠ μ2 - Медианы первой и второй выборок принадлежат одной генеральной совокупности    

# Воспользуемся функцией 
stat_value, p_value = stats.mannwhitneyu(x1, y1)
print(stat_value, p_value)
```

```python
# Примем уровень статистической значимости 0,05 
# Т.к. полученное значение p-value больше принятого уровня статистической значимости, 
# мы не можем отвергнуть нулевую гипотезу. 
# следовательно, согласно критерия Манна-Уитни между выборками отутствуют статистически значимые различия
```

## Задача 2
Исследовалось влияние препарата на уровень давления пациентов.

Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут.

Есть ли статистически значимые различия?

1е измерение до приема препарата: 150, 160, 165, 145, 155

2е измерение через 10 минут: 140, 155, 150, 130, 135

3е измерение через 30 минут: 130, 130, 120, 130, 125


```python
# Для зависимых выборок и повторных измерений применяется критерий Фридмана
```

```python
m1 = np.array([150, 160, 165, 145, 155])
m2 = np.array([140, 155, 150, 130, 135])
m3 = np.array([130, 130, 120, 130, 125])
```

```python
stat_value, p_value = stats.friedmanchisquare(m1, m2, m3)
print(stat_value, p_value)
```

```python
# Примем уровень статистической значимости 0,05 
# Т.к. полученное значение p-value меньше принятого уровня статистической значимости, 
# мы отвергаем нулевую гипотезу. 
# следовательно, согласно критерия Фридмана между выборками существуют статистически значимые различия
```

## Задача 5
Заявляется, что партия изготавливается со средним арифметическим 2,5 см. Проверить данную гипотезу, если известно, что размеры изделий подчинены нормальному закону распределения. Объем выборки 10, уровень статистической значимости 5%
2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

Задачу решить вручную

```python
# Задано, что совокупность подчинена нормальному закону распределения.
# Стандартная ошибка генеральной совокупности не изввестна.
# Целесообразно провести проверку соответствует ли выборка с указанными параметрами
# генеральной совокупности со средним арифметическим равным 2.5 возможно на основании критерия Стьюдента
# формула доверительного интервала |Xген - Xсреднвыб| < delta => -delta <  Xген - Xсреднвыб < + delta =>
# Xсреднвыб - delta < Xген < Xсреднвыб + delta
# delta - отклонение для среднего определяется как:
# delta = ty * SE
# ty - Коэффициент доверия
# SE - стандартная ошибка среднего 
# SE = σ / √n
# ty - табличное значение критерия Стьюдента для заданного уровня статистической значимости
```

```python
x = np.array([ 2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34])
# рассчитаем среднее по выборке
mean_x = round(x.sum()/ x.size, 5)
print(f'Среднее по выборке {mean_x} см')
# рассчитаем выборочное СКО несмещенное
sigma_not_off = round((((x - mean_x)**2).sum() / (x.size - 1))**0.5, 5)
print(f'СКО по выборке {sigma_not_off} см')
# рассчитаем стандартную ошибку среднего по выборке
SE = sigma_not_off / x.size**0.5
print(f'Стандартня ошибка срденего по выборке {SE} см')
```

```python
# для уровня статистичекой значимости 0,95 (видимо все таки в задаче опечатка уровень значимости 0,95 при ошибке 0,05) 
# при количестве степеней свбободы n-1 (10 - 1 = 9)
ty = 2.262
```

```python
delta = ty * SE
print(delta)
print(f'Доверительный интервал [{mean_x - delta};{mean_x + delta}] см')
```

```python
# Доверительный интервал с вероятностью 95% захватывает срдеднее по генеральной совокупности.
# Следовательно выборка принадлежит генеральной совокупности со средним арифметическим 2.5 см
```
